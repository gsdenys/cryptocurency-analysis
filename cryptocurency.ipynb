# Cryptocurency Data Analysis

Provide some analysis about the price changemente through the last 20 years.

## Import Required Libraries

```python
import pandas_datareader.data as pdr
import matplotlib.pyplot as plt
import pandas as pd

#libraries to work with date and time
from datetime import date
from datetime import timedelta
```

### Select the Data Range

```python
#Define the data source
source = 'yahoo'

rolling = 100
#select the date range that you want to export. by default I've chosen the 2000 
#first day to today
start = '2010-01-01'
end = date.today()
```

## Load Downloaded data

```python
#Download data
data = {}
data['BTC'] = pdr.DataReader('BTC-USD', source, start, end)
data['ETH'] = pdr.DataReader('ETH-USD', source, start, end)
data['LTC'] = pdr.DataReader('LTC-USD', source, start, end)
```

## Understanding Cryptocurrency Quotation

In this section we want to understand the quotation of each cryptocurrency separately. It'll make simple saw their maximum, minimum, and the average price of each one.

### Bitcoin

```python
btc = data['BTC']
```

```python
btc.describe()
```

```python
fig, ax = plt.subplots(figsize=(16,9))
ax.plot(btc.index, btc['Adj Close'])

ax.set_xlabel('Date')
ax.set_ylabel('Adjusted closing price ($)')
```

### Etherium

```python
eth = data['ETH']
```

```python
eth.describe()
```

```python
fig, ax = plt.subplots(figsize=(16,9))
ax.plot(eth.index, eth['Adj Close'])

ax.set_xlabel('Date')
ax.set_ylabel('Adjusted closing price ($)')
```

### Litecoin

```python
ltc = data['LTC']
```

```python
ltc.describe()
```

```python
fig, ax = plt.subplots(figsize=(16,9))
ax.plot(ltc.index, ltc['Adj Close'])

ax.set_xlabel('Date')
ax.set_ylabel('Adjusted closing price ($)')
```

### Comparative between them

```python
fig, ax = plt.subplots(figsize=(16,9))

ax.plot(btc.index, btc['Adj Close'], label='Bitcoin')
ax.plot(eth.index, eth['Adj Close'], label='Etherium')
ax.plot(ltc.index, ltc['Adj Close'], label='Litecoin')

ax.set_xlabel('Date')
ax.set_ylabel('Adjusted closing price ($)')
ax.legend()
```
